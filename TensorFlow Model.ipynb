```
# Creating Newral Network
x = tf.placeholder(dtype=tf.float32,shape=[None,10])
w1 = tf.Variable(initial_value=tf.zeros([10,100],dtype=tf.float32))
layer1 = tf.matmul(x,w1)
# Adding bias
b1 = tf.Variable(initial_value=tf.ones([100],dtype=tf.float32))
h1 = tf.sigmoid(tf.add(layer1,b1))

# Hidden Layer2
w2 = tf.Variable(initial_value=tf.zeros([100,100],dtype=tf.float32))
layer2 = tf.matmul(h1,w2)
# Adding bias
b2 = tf.Variable(initial_value=tf.ones([100],dtype=tf.float32))
h2 = tf.sigmoid(tf.add(layer2,b2))

# Output Later
w3 = tf.Variable(initial_value=tf.zeros([100,4],dtype=tf.float32))
layer3 = tf.matmul(h2,w3)
# Softmax Function
output = tf.nn.softmax(layer3)
```

```
output
```

```
# Backward Pass
y_true = tf.placeholder(dtype=tf.float32,shape=[None,4])

# Creating Loss function
loss = tf.nn.softmax_cross_entropy_with_logits_v2(labels=y_true,logits=output)
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
```

```
train = optimizer.minimize(loss)
```

```
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    epoch = 100
    for i in range(epoch):
        sess.run(train,feed_dict={x:x_train,y_true:y_train})
        
        cost = sess.run(loss,feed_dict={x:x_train,y_true:y_train})
        print('Epoch: %s, Loss %s'%(i,cost))
```
