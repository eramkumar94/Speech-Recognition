```
x_train = data['arr_0']
x_test = data['arr_1']
y_train = data['arr_2']
y_test = data['arr_3']
```

```
# Neural Network

#namescope
with tf.name_scope(name='placeholders'):
  inputx = tf.placeholder(dtype=tf.float32,shape=[None,10])
  inputy = tf.placeholder(dtype=tf.float32,shape=[None,4])
  
with tf.name_scope(name='weights'):
  w1 = tf.Variable(initial_value=tf.zeros([10,100],dtype=tf.float32))
  w2 = tf.Variable(initial_value=tf.zeros([100,100],dtype=tf.float32))
  w3 = tf.Variable(initial_value=tf.zeros([100,4],dtype=tf.float32))
  
  # bias
  b1 = tf.Variable(initial_value=tf.ones([100],dtype=tf.float32))
  b2 = tf.Variable(initial_value=tf.ones([100],dtype=tf.float32))
  b3 = tf.Variable(initial_value=tf.ones([4],dtype=tf.float32))

  
with tf.name_scope(name='connection_1'):
  # layer - 1 : 10 - 100 -> activation = sigmoid
  layer1 = tf.add(tf.matmul(inputx,w1),b1,name='layer_1')
  layer1_activation = tf.nn.relu(layer1)
  
with tf.name_scope(name='connection_2'):
  # layer -2 : 100 , 100 -> activation = sigmoid
  layer2 = tf.add(tf.matmul(layer1_activation,w2),b2,name='layer_2')
  layer2_activation = tf.nn.relu(layer2)
  
with tf.name_scope(name='output_layer'):
  y_ = tf.add(tf.matmul(layer2_activation,w3),b3,name='layer_3')
  y_actual = tf.sigmoid(y_,name='output')
```

# Saving Model

```
init = tf.global_variables_initializer()

with tf.Session() as sess:
  writer = tf.summary.FileWriter('./logs',graph=sess.graph)
  sess.run(init)
  writer.close()
```

```
# Compling model
# loss function

loss = tf.losses.softmax_cross_entropy(onehot_labels=inputy,logits=y_actual) # cross entropy
train = tf.train.AdamOptimizer(learning_rate=0.1).minimize(loss)
```

```
init = tf.global_variables_initializer()

with tf.Session() as sess:
  sess.run(init)
  writer = tf.summary.FileWriter('./logs_train',graph=sess.graph)
  for i in range(50):

    sess.run(train,feed_dict={inputx:x_train,inputy:y_train})    
    cost = sess.run(loss,feed_dict={inputx:x_train,inputy:y_train})
    print("EPOCH: %s, LOSS: %0.2f"%(i,cost))
    
    
  writer.close()
```
